```python
# functions to clean latitude and longitude values
def clean_latitude(value):
    import pandas as pd
    if pd.isna(value):
        return pd.NA
    
    value = str(value).strip().replace('"', '')
    
    if value in ['N/A', 'NA', '']:
        return pd.NA
    
    # normal decimal format with comma
    if ',' in value and value.count('.') <= 1:
        value = value.replace(',', '.')
        return pd.to_numeric(value, errors='coerce')
    
    # malformed format with many dots
    if value.count('.') > 1:
        digits = value.replace('.', '').replace(',', '').replace('-', '').strip()
        if digits == '':
            return pd.NA
        # latitude in southern Spain should start with 36, 37 or 38
        repaired = digits[:2] + '.' + digits[2:]
        return pd.to_numeric(repaired, errors='coerce')
    
    return pd.to_numeric(value, errors='coerce')


def clean_longitude(value):
    import pandas as pd
    if pd.isna(value):
        return pd.NA
    
    value = str(value).strip().replace('"', '')
    
    if value in ['N/A', 'NA', '']:
        return pd.NA
    
    # normal decimal format with comma
    if ',' in value and value.count('.') <= 1:
        value = value.replace(',', '.')
        return pd.to_numeric(value, errors='coerce')
    
    # malformed format with many dots
    if value.count('.') > 1:
        sign = -1 if '-' in value else 1
        digits = value.replace('.', '').replace(',', '').replace('-', '').strip()
        if digits == '':
            return pd.NA
        # longitude in southern Spain should start with 5, 6 or 7
        repaired = digits[:1] + '.' + digits[1:]
        repaired = pd.to_numeric(repaired, errors='coerce')
        return repaired * sign
    
    return pd.to_numeric(value, errors='coerce')
```

```python
# rebuild cleaned coordinate columns from the original raw columns
df['latitude_clean'] = df['felling_latitude_WGS84'].apply(clean_latitude)
df['longitude_clean'] = df['felling_longitude_WGS84'].apply(clean_longitude)
```
