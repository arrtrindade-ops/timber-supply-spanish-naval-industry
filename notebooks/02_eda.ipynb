```python
province_municipality_source_trees = (
    df.groupby(
        [
            'province',
            'municipality',
            'felling_toponymy_source'
        ]
    )['tree_quantity']
    .sum()
    .reset_index()
)

province_totals = (
    province_municipality_source_trees.groupby('province')['tree_quantity']
    .sum()
    .reset_index()
    .rename(columns={'tree_quantity': 'province_total'})
)

municipality_totals = (
    province_municipality_source_trees.groupby(['province', 'municipality'])['tree_quantity']
    .sum()
    .reset_index()
    .rename(columns={'tree_quantity': 'municipality_total'})
)

province_municipality_source_trees = province_municipality_source_trees.merge(
    province_totals,
    on='province'
)

province_municipality_source_trees = province_municipality_source_trees.merge(
    municipality_totals,
    on=['province', 'municipality']
)

province_municipality_source_trees = province_municipality_source_trees.sort_values(
    by=[
        'province_total',
        'municipality_total',
        'tree_quantity'
    ],
    ascending=[False, False, False]
)

province_municipality_source_trees
```

## 4. Spatio-temporal distribution of tree quantities

```python
period_province_municipality_trees = (
    df.groupby(
        [
            'felling_period',
            'province',
            'municipality'
        ]
    )['tree_quantity']
    .sum()
    .reset_index()
)

period_totals = (
    period_province_municipality_trees.groupby('felling_period')['tree_quantity']
    .sum()
    .reset_index()
    .rename(columns={'tree_quantity': 'period_total'})
)

province_totals_within_period = (
    period_province_municipality_trees.groupby(['felling_period', 'province'])['tree_quantity']
    .sum()
    .reset_index()
    .rename(columns={'tree_quantity': 'province_total'})
)

period_province_municipality_trees = period_province_municipality_trees.merge(
    period_totals,
    on='felling_period'
)

period_province_municipality_trees = period_province_municipality_trees.merge(
    province_totals_within_period,
    on=['felling_period', 'province']
)

period_province_municipality_trees = period_province_municipality_trees.sort_values(
    by=[
        'period_total',
        'province_total',
        'tree_quantity'
    ],
    ascending=[False, False, False]
)

period_province_municipality_trees
```

```python
period_province_municipality_trees = (
    df.groupby(
        [
            'felling_period',
            'province',
            'municipality'
        ]
    )['tree_quantity']
    .sum()
    .reset_index()
)

province_totals_within_period = (
    period_province_municipality_trees.groupby(['felling_period', 'province'])['tree_quantity']
    .sum()
    .reset_index()
    .rename(columns={'tree_quantity': 'province_total'})
)

period_province_municipality_trees = period_province_municipality_trees.merge(
    province_totals_within_period,
    on=['felling_period', 'province']
)

period_province_municipality_trees = period_province_municipality_trees.sort_values(
    by=[
        'felling_period',
        'province_total',
        'tree_quantity'
    ],
    ascending=[True, False, False]
)

period_province_municipality_trees
```

Records by year

```python
period_summary = (
    df[['felling_period', 'tree_quantity']]
    .dropna()
    .copy()
)

# extrai o primeiro ano que aparece no texto do período
period_summary['period_start_year'] = (
    period_summary['felling_period']
    .astype(str)
    .str.extract(r'(\d{4})')[0]
    .astype(float)
)

# remove linhas onde não conseguiu encontrar ano
period_summary = period_summary.dropna(subset=['period_start_year']).copy()

# soma quantidades por período
period_summary = (
    period_summary
    .groupby(['felling_period', 'period_start_year'], as_index=False)['tree_quantity']
    .sum()
    .sort_values(by='period_start_year', ascending=True)
)

plt.figure(figsize=(12, 6))
plt.bar(period_summary['felling_period'], period_summary['tree_quantity'])
plt.title('Tree quantities by felling period')
plt.xlabel('Felling period')
plt.ylabel('Tree quantity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Final analytical summary

- Cádiz is by far the province with the highest total quantity of recorded felled trees, followed by Málaga, Sevilla and Huelva.
- The municipalities with the largest quantities of recorded felled trees are Cortes de la Frontera, Jimena de la Frontera, Jerez de la Frontera and Alcalá del Valle.
- Tree felling activity is strongly concentrated in a relatively small number of municipalities and specific places.
- Oak, holm oak and gall oak are the dominant tree species in the dataset and together account for the majority of recorded tree quantities.
- Pine is also significant, although much less represented than the main oak-related species.
- Comunal landownership is the most important form of land control represented in the dataset, followed by noble estates, private land and ecclesiastical land.
- A relatively small number of landowners account for a very large proportion of the total quantity of recorded felled trees.
- Among the most important landowners are Jerez de la Frontera, Jimena de la Frontera, the Duque de Arcos, the Marqués de Benamejí and the Duque de Medina Sidonia.
- The periods 1749–1750 and 1748–1749 are the most productive phases represented in the dataset.
- A substantial proportion of the records includes valid coordinates, making it possible to create point maps of felling locations and combine them with polygon maps of provinces and municipalities.
